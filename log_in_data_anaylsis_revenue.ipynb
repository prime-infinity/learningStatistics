```python
import pandas as pd
```

```python
data = {'company': ['Tesla', 'UBER', 'Amazon', 'Jindal Steel', 'Axis Bank', 'Vedanta'],
        'revenue': [31, 11, 386, 4.7, 5.6, 11.3]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.plot(x="company",y="revenue",kind="bar")
```

# using log function to squish the bar

```python
df.plot(x="company",y="revenue",kind="bar",logy=True)
```
