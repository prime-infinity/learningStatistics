# using normal z score to detect outliers

```python
df.revenue_mln.describe() #we are dealing with price only
```

```python
_, mean, std, *_ = df.revenue_mln.describe()
```

```python
mean,std
```

```python
def get_z_score(value,mean,std):
    return (value-mean)/std
```

```python
#create another column in the dataframe called zscore and store it
```

```python
df["z_score"] = df.revenue_mln.apply(lambda x: get_z_score(x,mean,std))
```

```python
df.head()
```

```python
#outliers based on zscore
df[df.z_score < 3]
```

```python
#but as you can see, many valuable and legit datasets has been removed
#this is not a bad value to remove unless it is altering the values of the rest of our dataset, there are other values that may
#be outliers, but because this values affects the mean more than them, we might not be detecting them
```

# usnig modified zscore to detect outliers

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s - median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(df.revenue_mln)
median = np.median(df.revenue_mln)
MAD,median
```

```python
def get_modified_z_score(x,median,MAD):
    return 0.6745 *(x-median)/MAD
```

```python
get_modified_z_score(2787.965087,median,MAD)
```

```python
df["mod_z_score"] = df.revenue_mln.apply(lambda x: get_modified_z_score(x,median,MAD))
df.head()
```

```python
#outliers based on mod_z_score
df[df.mod_z_score>3.5]
```

```python
df[df.mod_z_score<=3.5] #outliers wth modified zscore
```
