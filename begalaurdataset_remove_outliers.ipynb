```python
import pandas as pd
import seaborn as sn
```

```python
df = pd.read_csv("Bengaluru_House_Data.csv")
df
```

```python
sn.histplot(df.price, kde=True)
```

```python
df.price.describe()
```

```python
mean = df.price.mean()
mean
```

```python
df.price.describe()
```

```python
min_quant, max_quant = df.price.quantile([0.01,0.99])
min_quant, max_quant
```

```python
df_no_outier = df[(df.price>min_quant) & (df.price<max_quant)]
df_no_outier
```

```python
std = df.price.std()
```

```python
lower_bound = mean - 4*std
upper_bound = mean + 4*std
```

```python
df_no_outliers = df[(df.price >= lower_bound) & (df.price <= upper_bound)]
df_no_outliers
```

```python
sn.histplot(df_no_outliers.price, kde=True)
```
