```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#find income dataset
data = {'income': [5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000,
                   55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000, 95000, 100000,
                   105000, 110000, 115000, 120000, 125000, 130000, 135000, 140000, 145000,
                   150000, 155000, 160000, 165000, 170000, 175000, 180000, 185000, 190000,
                   195000, 200000],
        'count': [4371, 3295, 5825, 6047, 6097, 5738, 6100, 5720, 5098, 4991,
                  5152, 4194, 4411, 3709, 3811, 3766, 3565, 3035, 2753, 2644,
                  2692, 2180, 2278, 1932, 1992, 1656, 1547, 1424, 1342, 1134,
                  1489, 1006, 1085, 784, 869, 849, 730, 688, 697, 508]}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
g.set(xscale="log");
```
